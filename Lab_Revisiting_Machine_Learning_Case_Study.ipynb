```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.stats.api as sms
import matplotlib as mpl
import seaborn as sns
from math import ceil 
```

```python
df = pd.read_csv('learningSet.csv')
df
```

```python
categorical = df.select_dtypes([object])
categorical
```

```python
categorical.isnull().all()
```

```python
count = categorical.isnull().sum()
count
```

```python
drop_list = []
```

```python
drop_list.append(['OSOURCE','ZIP'])
```

```python
drop_list
```

```python
categorical_copy = categorical.copy()
print(categorical_copy.columns)
```

```python

columns_to_drop = [col[0] for col in drop_list]
categorical_copy = categorical_copy.drop(columns=columns_to_drop)
categorical_copy
```

```python
for col in categorical_copy.columns:
    pct_missing = np.mean(categorical_copy[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python

for col in categorical_copy.columns:
    pct_missing = np.mean(categorical_copy[col].isnull())
    if pct_missing > 0.85:
        print('{} - {}%'.format(col, round(pct_missing * 100)))
```

```python
# No missing value , so we do not need to drop any column
```

```python
print(categorical_copy['GENDER'].value_counts())
categorical_copy['GENDER'] = categorical_copy['GENDER'].fillna('F')
```
